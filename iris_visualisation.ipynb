```python
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = load_iris()
```

```python
X=data.data
y=data.target
```

```python
ydf=pd.DataFrame(y,columns=["class"])
```

```python
xdf=pd.DataFrame(X,index=np.arange(0,150),columns=data.feature_names)
```

```python
df=pd.concat([xdf,ydf],axis=1)
```

```python
df.head()
```

```python
df.columns = ['sepal_length', 'sepal_width',"petal_length","petal_width","class"]
```

```python
df.head()
```

```python
plt.scatter(df.index,df["sepal_width"])
```

```python
sns.scatterplot(x=df.index,y=df["sepal_width"],hue=y)
```

```python
df.groupby("class").mean()
```

### 3.  STRIP PLOT

```python
sns.stripplot(y=df["sepal_width"])
```

```python
sns.stripplot(y=df["sepal_width"],x=df["class"])
```

### Swarm Plot

```python
sns.swarmplot(x=df["sepal_width"])
```

```python
sns.swarmplot(y=df["sepal_width"],x=df["class"])
```

## Uni-variate summary plots :


### HISTOGRAMS 

```python
plt.hist(df["petal_width"])
```

```python
sns.distplot(df["petal_width"],kde=False,bins=10)
```

### Density Plot

```python
plt.figure(figsize=(5,5))
df["petal_length"].plot(kind="density")
```

```python
sns.set(rc={"figure.figsize":(5,5)})
sns.kdeplot(df["petal_length"],shade=True)
```

```python
plt.violinplot(df.values,showmedians=True)
plt.show()
```

```python
sns.violinplot(y=df["petal_width"],x=df["class"],orient="vertical")
```

### BAR

```python
df["class"].value_counts().plot.bar()
```

```python
sns.countplot(df["class"])
```

### Pie Chart

```python
dfsample=df.sample(frac=0.75) #! verilerin yzde 75ini rastgele bir ekilde aldk
```

```python
plt.pie(dfsample["class"].value_counts(),labels=["Setosa","Versicolor","Virginica"],autopct="%.3f",shadow=True)
plt.show()
```

```python
plt.plot(df.index,df["petal_length"],markevery=1,marker="d")
for name,group in df.groupby("class"):
    plt.plot(group.index,group["petal_length"],label=name,markevery=1,marker="d")
plt.legend()
plt.show()
```
